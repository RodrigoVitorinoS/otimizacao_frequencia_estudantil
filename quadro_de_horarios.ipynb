```python
#Importando bibliotecas
import pyomo.environ as pyo
from pyomo.opt import SolverStatus, TerminationCondition

#Pesos
pesos_neja_3= {'Geografia': 15.71,
               'Matemática': 14.92,
               'Português': 15.23,
               'História': 13.72,
               'Filosofia': 14.06,
               'Educação Física': 13.96,
               'Sociologia': 12.4}

materias_neja_3 = list(pesos_neja_3.keys())
pesos_neja_2= {'Matemática': 14.41,
               'Português': 15.08,
               'Vida': 15.14,
               'Química': 13.22,
               'Biologia': 13.06,
               'Física': 15.15,
               'Artes': 13.94}

materias_neja_2 = list(pesos_neja_2.keys())
pesos_neja_1= {'Geografia': 13.59,
               'Matemática': 16.47,
               'Português': 16.21,
               'História': 15.24,
               'Filosofia': 13.12,
               'Educação Física': 13.12,
               'Inglês': 12.24}

materias_neja_1 = list(pesos_neja_1.keys())
pesos_terceiro= {'Geografia': 9.33,
                 'Matemática': 9.39,
                 'Português': 9.73,
                 'História': 7.03,
                 'Filosofia': 9.3,
                 'Química': 8.79,
                 'Biologia': 9.39,
                 'Educacao Física': 9.33,
                 'Física': 9.25,
                 'Língua Estrangeira': 9.6,
                 'Sociologia': 8.87}

materias_terceiro = list(pesos_terceiro.keys())
pesos_segundo= {'Geografia': 8.83,
                'Matemática': 8.52,
                'Português': 8.58,
                'História': 8.84,
                'Filosofia': 8.44,
                'Química': 8.68,
                'Biologia': 6.64,
                'Educação Física': 8.51,
                'Física': 8.83,
                'Artes': 8.73,
                'Língua Estrangeira': 6.89,
                'Sociologia': 8.51}

materias_segundo = list(pesos_segundo.keys())
pesos_primeiro = {
    'Geografia':9.3,
    'Matemática':9.44,
    'Português':9.31,
    'Projeto de Vida':8.44,
    'Química':9.15,
    'Educacao Física':9.2,
    'História':9.49,
    'Filosofia':9.3,
    'Biologia':9.48,
    'Inglês':7.6,
    'Física':9.28

}

materias_primeiro = list(pesos_primeiro.keys())
pesos_nono= {'Geografia': 9.97,
             'Matemática': 10.23,
             'Português': 10.09,
             'História': 10.22,
             'Educação Física': 10.22,
             'Inglês': 9.84,
             'Ciências': 10.01,
             'Artes': 9.6,
             'RPM': 9.82,
             'PT': 10.0}

materias_nono = list(pesos_nono.keys())
pesos_oitavo= {'Geografia': 10.01,
               'Matemática': 9.84,
               'Português': 9.9,
               'História': 9.97,
               'Educação Física': 10.17,
               'Inglês': 9.72,
               'Ciências': 10.05,
               'Letramento Matemática': 9.98,
               'Letramento Português': 10.2,
               'Artes': 10.15}

materias_oitavo = list(pesos_oitavo.keys())
pesos_setimo = {'Geografia': 9.98,
                'Matemática': 10.25,
                'Português': 9.9,
                'História': 9.93,
                'Educação Física': 10.24,
                'Inglês': 9.86,
                'Ciências': 9.8,
                'Letramento Matemática': 10.23,
                'Letramento Português': 10.37,
                'Artes': 9.43}

materias_setimo = list(pesos_setimo.keys())
pesos_sexto = {'Geografia': 9.89,
               'Matemática': 10.15,
               'Português': 10.27,
               'História': 10.13,
               'Educação Física': 9.88,
               'Inglês': 10.02,
               'Ciências': 9.95,
               'Letramento Matemática': 10.1,
               'Letramento Português': 9.76,
               'Artes': 9.85}

materias_sexto = list(pesos_sexto.keys())
materias_por_ano = {
    'Sexto': materias_sexto,
    'Sétimo': materias_setimo,
    'Oitavo': materias_oitavo,
    'Nono': materias_nono,
    'Primeiro': materias_primeiro,
    'Segundo': materias_segundo,
    'Terceiro': materias_terceiro,
    'Neja 1': materias_neja_1,
    'Neja 2': materias_neja_2,
    'Neja 3': materias_neja_3

}
pesos_por_ano = {
    'Sexto': pesos_sexto,
    'Sétimo': pesos_setimo,
    'Oitavo': pesos_oitavo,
    'Nono': pesos_nono,
    'Primeiro': pesos_primeiro,
    'Segundo': pesos_segundo,
    'Terceiro': pesos_terceiro,
    'Neja 1': pesos_neja_1,
    'Neja 2': pesos_neja_2,
    'Neja 3': pesos_neja_3

}

#Dias e tempos
dias = [
      'Segunda',
      'Terca',
      'Quarta',
      'Quinta',
      'Sexta'
]
tempos = [
1,2,3,4,5
]


# SOLVER
def resolver_modelo(modelo,materias, log = False, gap=None, tempo_limite = None ):

        opt = pyo.SolverFactory('glpk')

        if gap is not None:
            opt.options['mipgap'] = gap
        if tempo_limite is not None:
            opt.options['tmlim'] = tempo_limite


        resultado = opt.solve(modelo, tee = log)


        if resultado.solver.status == SolverStatus.ok :
          for i in range(5):
            expr = 0

            for j in modelo.aulas:
              if pyo.value(modelo.aulas[j]) < 0.5:

              expr += modelo.aulas[j]
              else:
              expr += (1 - modelo.aulas[j])



            modelo.c.add( expr >= 1 )
            results = opt.solve(modelo, tee= log)
            print('\n')

            display(HTML(f"<h2>QUADRO {i+1}</h2>"))
            materias_na_semana = {}
            for dia in dias:
              materias_no_dia = []
              for materia in materias:
                for tempo in tempos:
                  valor = round(pyo.value(modelo.aulas[dia, materia, tempo]), 3)
                  if valor>0:
                      materias_no_dia.append(materia)
              materias_na_semana[dia] = materias_no_dia


                    # Verificar o comprimento das listas no dicionário
            comprimentos = {chave: len(valor) for chave, valor in materias_na_semana.items()}

            # Encontrar o comprimento máximo
            comprimento_maximo = max(comprimentos.values())

            # Preencher as listas mais curtas com valores nulos (NaN)
            for chave, valor in materias_na_semana.items():
                if len(valor) < comprimento_maximo:
                    materias_na_semana[chave] += [''] * (comprimento_maximo - len(valor))
            quadro_horario = pd.DataFrame(materias_na_semana)
            display(quadro_horario)


        elif resultado.solver.status == SolverStatus.warning:
            print('Atenção: Problema pode não ter solução ótima')
        elif resultado.solver.termination_condition == TerminationCondition.infeasible:
            print('O modelo é inviável. Não há solução possível.')
        else:
            print(f'Solver status: {resultado.solver.status}')

        print(resultado.solver.status)
        print('Solução  encontrada!')
        print(f'Função objetivo: {pyo.value(modelo.obj)}\n')

# MODELO
def criar_modelo_inteiro(materias, dias, tempos_materia , pesos, tempos):
        tempos_dia = 6
        modelo = pyo.ConcreteModel()

        modelo.materias = pyo.Set(initialize = materias)
        modelo.dias = pyo.Set(initialize = dias)
        modelo.tempos= pyo.Set(initialize = tempos)

        modelo.tempos_dia = pyo.Param(initialize = tempos_dia)
        modelo.tempos_materia = pyo.Param(modelo.materias, initialize = tempos_materia)
        modelo.pesos = pyo.Param(modelo.materias, initialize =pesos)


        modelo.aulas = pyo.Var(modelo.dias, modelo.materias, modelo.tempos, domain = pyo.Binary)
        modelo.menor_qtd_materias = pyo.Var(within= pyo.NonNegativeIntegers)

        for dia in dias:
          for materia in tempos_materia.keys():
            for i in range(tempos_materia[materia]+1, 6):
              modelo.aulas[dia, materia, i].fix(0)


        modelo.obj = pyo.Objective(
            expr =  modelo.menor_qtd_materias,
            sense = pyo.maximize
        )

        modelo.restr_menor_qtd_aulas = pyo.Constraint(
            modelo.dias,
            rule = lambda modelo, dia: sum(modelo.pesos[materia]*modelo.aulas[dia, materia, tempo]
                                          for materia in materias
                                          for tempo in tempos ) >= modelo.menor_qtd_materias

        )

        modelo.restr_tempos_dia = pyo.Constraint(
            modelo.dias,
            rule = lambda modelo, dia: sum( modelo.aulas[dia,materia, tempo]
                                          for materia in materias
                                            for tempo in tempos ) <= modelo.tempos_dia
        )


        modelo.c = pyo.ConstraintList()

        modelo.restr_tempos_materia = pyo.Constraint(
            modelo.materias,
            rule = lambda modelo, materia:
            sum(modelo.aulas[dia, materia, tempo] for dia in dias for tempo in tempos) == modelo.tempos_materia[materia]
        )


        for materia in materias:

          modelo.add_component(f'restr_mesmos_tempos_{materia}', pyo.Constraint(
              [tempo for tempo in range(1, tempos_materia[materia]+1)],
              rule = lambda modelo,tempo: sum(modelo.aulas[dia, materia, tempo] for dia in dias) == 1
          ))

          # if tempos_materia[materia] >= 3:
          modelo.add_component(f'restr_duplas_1_{materia}', pyo.Constraint(
            modelo.dias,
            rule=lambda modelo, dia: modelo.aulas[dia, materia, 2]<= modelo.aulas[dia, materia, 1]))



          # if tempos_materia[materia] >= 4:
          modelo.add_component(f'restr_duplas_2_{materia}', pyo.Constraint(
            modelo.dias,
            rule=lambda modelo, dia: modelo.aulas[dia, materia, 4]<= modelo.aulas[dia, materia, 3]))

          # if tempos_materia[materia] == 5:
          modelo.add_component(f'restr_trios_{materia}', pyo.Constraint(
            modelo.dias,
            rule=lambda modelo, dia: modelo.aulas[dia, materia, 5]<= modelo.aulas[dia, materia, 4]))

          modelo.add_component(f'restr_max_por_dia_{materia}', pyo.Constraint(
            modelo.dias,
            rule=lambda modelo, dia: sum(modelo.aulas[dia, materia, tempo] for tempo in tempos)<= 4))


        return modelo
```
